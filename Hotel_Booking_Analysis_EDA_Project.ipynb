## 10. What is the average daily rate (ADR) for different meal types, and how does it vary among guests who choose different meal plans?

```python
# chart 10
# Group data by 'meal' and calculate mean ADR for each meal type
meal_adr = df.groupby('meal')['adr'].mean()

# Define meal types and their corresponding labels for visualization
meal_labels = {
    'BB': 'Bed & Breakfast',
    'HB': 'Half board',
    'FB': 'Full board',
    'SC': 'Self catering',
    'Undefined': 'Undefined'
}

# Convert meal types to labels for better visualization
meal_adr.index = meal_adr.index.map(meal_labels)

# Create a bar plot to visualize average daily rate for different meal types
plt.figure(figsize=(10, 6))
meal_adr.plot(kind='bar')
plt.xlabel('Meal Type')
plt.ylabel('Average Daily Rate (ADR)')
plt.title('Average Daily Rate for Different Meal Types')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#show
meal_adr
```

- Mostly people chose Full board and half board.

## 11. Are there seasonal patterns in booking cancellations based on the arrival date (year, month, and day of the week)?

```python
#chart 11
# Select the relevant columns
data = df[['arrival_date_year', 'arrival_date_month', 'arrival_date_week_number', 'is_canceled']]

# Group the data by year, month, and day of the week, and calculate the cancellation rate
cancellation_rate = data.groupby(['arrival_date_year', 'arrival_date_month', 'arrival_date_week_number'])['is_canceled'].mean().reset_index()

# Create a line plot for seasonal patterns by month
plt.figure(figsize=(10, 6))
sns.lineplot(data=cancellation_rate, x='arrival_date_month', y='is_canceled', hue='arrival_date_year', marker='o')
plt.title('Seasonal Patterns of Booking Cancellations by Month')
plt.xlabel('Month')
plt.ylabel('Cancellation Rate')
plt.legend(title='Year')
plt.xticks(rotation=45)
plt.grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Create a line plot for seasonal patterns by day of the week
plt.figure(figsize=(10, 6))
sns.lineplot(data=cancellation_rate, x='arrival_date_week_number', y='is_canceled', hue='arrival_date_year', marker='o')
plt.title('Seasonal Patterns of Booking Cancellations by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Cancellation Rate')
plt.legend(title='Year')
plt.grid(True)
```

- If we see monthly wise cancellation so we can see that there is some trend of cancellation like in january, February, March and April most of people canceled there booking
- we can see that data for 2016 and 2017 most of people canceled their booking in this 4 months.

## 12. Is there any relationship between the lead time and is_canceled variable? If the the lead time increases the cancellation rate is also increases ?

```python
#chart 12
plt.figure(figsize=(10, 6))
data = df.groupby("lead_time")["is_canceled"].describe()
sns.regplot(x=data.index,y=data["mean"]*100)
plt.title("Lead Time vs is_canceled correlation")
plt.xlabel("Lead Time")
plt.ylabel("is_canceled")
plt.show()
```

- We can see here there is positive correlation between is_canceled and Lead Time columns
- It means that as the lead time (the number of days between booking and arrival) increases, the likelihood of a reservation being canceled also increases. In other words, longer lead times are associated with a higher probability of cancellations.
