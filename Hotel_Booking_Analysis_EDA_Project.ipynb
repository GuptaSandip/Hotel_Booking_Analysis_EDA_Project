## 1.How does the "is_canceled" variable relate to "market_segment"? Are there certain segments more likely to cancel bookings?

```python
df[['is_canceled','market_segment']].head(10)
```

```python
df.market_segment.unique()
```

####Above we can see that in market segment there is different values or say different segment for booking hotel.
####So, we are trying to define first all this segment one by one because it help us on visualization time to gain insights.
- 1. Direct - The booking made directly with the hotel.
- 2. Corporate - The booking was made by a corparate entity or for the business purpose.
- 3. Online TA - The booking made through an online agency
- 4. Offline TA/TO - The booking made through an offline travel agency/tour operator.
- 5. Complementary - The booking is complementary, often associated with loyalty programs or special arrangment.
- 6. Groups - The booking is part of a group reservation.
- 7. Aviation - It represents bookings or reservation made by or through aviaion-related sources.(bookings made by may be airlines company or any aviation industry)  

```python
# we are choosing stacked bar plot wich gives us every segment barplot like a group.
plt.figure(figsize=[10,5])
segment_cancel_rate = df.groupby('market_segment')['is_canceled'].mean().reset_index()

sns.barplot(x='market_segment',y='is_canceled',data=segment_cancel_rate)
plt.show()
```

```python
ct = pd.crosstab(df['market_segment'],df['is_canceled'],normalize='index')
ct.plot(kind='bar',stacked=True)
```
