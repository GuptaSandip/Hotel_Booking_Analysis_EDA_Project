####Above we can see that in market segment there is different values or say different segment for booking hotel.
####So, we are trying to define first all this segment one by one because it help us on visualization time to gain insights.
- 1. Direct - The booking made directly with the hotel.
- 2. Corporate - The booking was made by a corparate entity or for the business purpose.
- 3. Online TA - The booking made through an online agency
- 4. Offline TA/TO - The booking made through an offline travel agency/tour operator.
- 5. Complementary - The booking is complementary, often associated with loyalty programs or special arrangment.
- 6. Groups - The booking is part of a group reservation.
- 7. Aviation - It represents bookings or reservation made by or through aviaion-related sources.(bookings made by may be airlines company or any aviation industry)  

```python
# we are choosing stacked bar plot wich gives us every segment barplot like a group.
plt.figure(figsize=[10,5])
segment_cancel_rate = df.groupby('market_segment')['is_canceled'].mean().reset_index()

sns.barplot(x='market_segment',y='is_canceled',data=segment_cancel_rate)
# plt.legend(title='Booking Cancel',labels=['not cancel','cancel'])
plt.xlabel("Market Segment")
plt.ylabel("Cancellation Rate")
plt.title("Cancellation Rate by Market Segment")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have the data in a pandas DataFrame named 'df'
# Replace 'adr' and 'reserved_room_type' with the actual column names in your dataset

# Calculate the average adr for each reserved_room_type
avg_adr_by_room_type = df.groupby('reserved_room_type')['adr'].mean().sort_values(ascending=False)

# Create a bar plot using seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.barplot(x=avg_adr_by_room_type.index, y=avg_adr_by_room_type.values)
plt.xlabel('Reserved Room Type')
plt.ylabel('Average Daily Rate (ADR)')
plt.title('Average Daily Rate by Reserved Room Type')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset into a DataFrame
# df = pd.read_csv('hotel_bookings.csv')  # Replace with your data loading code

# Group data by 'meal' and calculate mean ADR for each meal type
meal_adr = df.groupby('meal')['adr'].mean()

# Define meal types and their corresponding labels for visualization
meal_labels = {
    'BB': 'Bed & Breakfast',
    'HB': 'Half board',
    'FB': 'Full board',
    'SC': 'Self catering',
    'Undefined': 'Undefined'
}

# Convert meal types to labels for better visualization
meal_adr.index = meal_adr.index.map(meal_labels)

# Create a bar plot to visualize average daily rate for different meal types
plt.figure(figsize=(10, 6))
meal_adr.plot(kind='bar' )#color='skyblue')
plt.xlabel('Meal Type')
plt.ylabel('Average Daily Rate (ADR)')
plt.title('Average Daily Rate for Different Meal Types')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
